```python
import random
```

```python
# 해당 학년과 특정 과목의 학년 전체의 평균점수, 최저점, 최고점, 만일 반 값도 주어져 있다면 반평균까지 구할 수 있는 함수
def get_grade_average(ginput,subinput,school,rinput=None):
    cal = []

    if ginput not in school:
        print("{} 학년은 없습니다. 다시 입력하세요.".format(ginput))
        return
                                                                 # 학년 입력값이 school에 있는 경우
    if subinput not in ["국어","영어","수학"]:
        print("{} 과목은 유효하지 않습니다. 다시 입력하세요.".format(subinput))                          # 과목 입력값이 국어, 영어, 수학이 아닐때
        return

    if rinput:
        if rinput not in school[ginput]:
            print("{} 은 없습니다. 다시 입력하세요.".format(rinput))
            return
        cal = [student[subinput] for student in school[ginput][rinput].values()]
        average = sum(cal)/len(cal)
        high = max(cal)
        low = min(cal)
        print("{}의 {} {} 과목 평균 점수는 {}점 입니다." .format(ginput,rinput,subinput,average))
        print("{}의 {} {} 과목 최고점은 {}, 최저점은 {}." .format(ginput,rinput, subinput,high,low))
        exit()

    else :
        for room in school[ginput].values():                                # 딕셔너리 school 자료형에 ginput라는 key를 가지는  value 들을 room에 저장함
            for student in room.values():                                   # 딕셔너리 room 자료형에 있는 value들을 student에 저장함
                if subinput in student:                                     # 조건문 : 만약 subinput이 student에 존재한다면 subinput에 따른 value 값들은 cal에 저장함
                    cal.append(student[subinput])
        average = sum(cal)/len(cal)
        high = max(cal)
        low = min(cal)
        print("{}의 {} 과목 평균 점수는 {}점 입니다." .format(ginput,subinput,average))
        print("{}의 {} 과목 최고점은 {}, 최저점은 {}." .format(ginput,subinput,high,low))
        exit()


    return
```

```python
# 학교의 학생들의 성적을 딕셔너리 자료형으로 생성
school = {}                                                                     # 학년 데이터를 저장할 school 초기화, 반복문을 돌며 학년 데이터를 저장
for grade in range(6):
    room = {}                                                                   # 반 데이터를 저장할  room 초기화, 반복문을 돌며 반 데이터를 저장
    for room_num in range(5):
        student = {}                                                            # 학생 데이터를 저장할 student 초기화, 반복문을 돌며 학생 데이터를 저장
        for stu_num in range(10):
            point = {                                                           # 학생의 국어, 영어, 수학 점수 생성
                '국어': random.randint(0, 100),
                '영어': random.randint(0, 100),
                '수학': random.randint(0, 100)
            }
            student["학생 " + str(stu_num + 1)] = point                         # 학생 key를 생성하고 랜덤 생성된 점수를 value로 저장
        room[str(room_num + 1) + "반"] = student                               # 반 key를 생성하고 학생데이터를  value로 저장
    school[str(grade + 1) + "학년"] = room                                      # 학년 key를 생성하고 반데이터를 value로 저장
                                                                                # 반복에 따라 새로운 데이터를 지속생성하게 해야 중복데이터가 생기지 않음

print(school["1학년"])
print(school["2학년"])
print(school["3학년"])
```

```python
# 학년과 과목으로 해당 학년의 과목 평균값을 구함
ginput = input("학년을 입력하세요:(1~6): ")+"학년"
subinput = input("과목을 입력하세요(국어, 영어, 수학): ")

get_grade_average(ginput,subinput,school)

rinput = input("반을 입력하세요:(1~5): ")+"반"

get_grade_average(ginput,subinput,school,rinput)
```
