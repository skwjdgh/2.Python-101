```python
# 1. Shallow copy & Deep copy
a = [1,2,3]
id(a)
```

```python
# Whey you put a to b, b has the same address like a
b=a
id(b)
print(b)
```

```python
a[0] = 50
id(b)
print(a) # b를 변경했는데 a까지 변경된 상황이 발생함. 객체의 값이 변했으므로
id(a) #주소값은 변하지 않음
```

```python
# 2. Shallow copy
c = a.copy()
id(c)
```

```python
print(c) # a의 값이 변했으나 c의 값은 변하지 않음
```

```python
#
print(id(2)) # literal can't have the other address
print(id(2))
print(id(2))
print(id(2))

print(id([1,2,3])) # unlike the litheral, datatype item always have other address
print(id([1,2,3]))
print(id([1,2,3]))
print(id([1,2,3]))
```

```python
ex_a = [1,2,[3,4,5]]
ex_b = [1,2,[3,4,5]]

print(id(ex_a[0]), id(ex_b[0]))
print(id(ex_a[1]), id(ex_b[1]))
print(id(ex_a[2]), id(ex_b[2]))
```

```python
ex_a_1 = [1,2,[3,4,5]]
ex_b_1 = ex_a_1.copy()

print(id(ex_a_1[0]), id(ex_b_1[0]))
print(id(ex_a_1[1]), id(ex_b_1[1]))
print(id(ex_a_1[2]), id(ex_b_1[2]))
```

```python
ex_a_1[0] = 10

print(id(ex_a_1[0]), id(ex_b_1[0]))
print(id(ex_a_1[1]), id(ex_b_1[1]))
print(id(ex_a_1[2]), id(ex_b_1[2]))
```
