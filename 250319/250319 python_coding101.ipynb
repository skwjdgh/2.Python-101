```python
def outer_function():
    global a #global을 통해 지역변수를 전역변수로 설정
    a = 20
    def inner_function():
        global a
        a = 30
        print('a =', a)
    inner_function()
    print('a =', a)
a = 10
outer_function()
print('a =', a)
```
