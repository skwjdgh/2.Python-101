```python
# 모듈 호출
import random
import os
```

```python
# 투구값 산출을 위한 함수 선언
def randthrow(rpin):
    pin = random.randint(0,rpin)
    return pin
```

```python
def points(fr,result,tr1,tr2,tr3,st):

    result[fr][0] = tr1
    result[fr][1] = tr2
    result[fr][2] = tr3
    result[fr][3] = tr1 + tr2 + tr3
    result[fr][4] = result[fr][4] + result[fr][3]
    result[fr][5] = st

    return result
```

```python
def pastfr(fr,result,th):
    for pframes in range(fr-1, fr+1):
        if pframes+1 < 0:
            continue
        else:
            pastframe = str(pframes+2) +" 프레임"
            result[pastframe][3] = result[pastframe][3] + th
            result[pastframe][4] = result[pastframe][4] + th
            if result[pastframe][5] > 0:
                result[pastframe][5] = result[pastframe][5] - 1

        return result
```
