```python
import random
```

```python
#필요한 상수 선언

defaultframe = 10 #플레이할 프레임은 10프레임
defaultpin = 10   #기본 핀은 10개

#상태값을 나타내는 상수들
normal = 0
spare = 1
strike = 2
```

```python
# 사용할 변수들을 선언한다
throw1 = 0  #첫번째 투구
throw2 = 0  #두번째 투구
throw3 = 0  #세번째 투구
sum = throw1 + throw2 + throw3 #투구들의 합계
total = 0   #누적합을 표현하기 위한 변수
statement = normal
```

```python
#데이터들을 누적할 자료형
tempresult = {}                                     # 프레임들의 연산값을 저장할 dictionary
for frame in range(defaultframe):                   # 디폴트 프레임의 크기에 따라 반복문 실행
    currentframe = str(frame+1) + "프레임"          # 키 값을 생성하는 반복문
    tempresult[currentframe] = [throw1, throw2,throw3, sum,total,statement]        # 반복문을 통해 데이터를 누적할 공간을 생성한다.
```

```python
# 함수 1 - 투구 값을 생성하는 함수
def randomthrow(a):
    res = random.randint(0,a)
    return res
```

```python
# 함수 2  - 범위를 설정하는 함수
def frames(b,c):
    fra = range(b)
    return fra
```

```python
#볼링게임
print("볼링게임을 시작합니다.")
for frame in range(defaultframe):                   # 헷갈리지 않게 동일 이름으로 하자

    print("{}프레임".format(frame+1))
    throw1 = randomthrow(defaultpin)                # 기본핀을 파라미터로 랜덤값을 만들자

    print("1투구 {} 점입니다.".format(throw1))

    print("1.", tempresult)
    if throw1 == defaultpin:                                 # 첫 투구 점수가 10일때, 스트라이크
        print("스트라이크\n")
        tempresult[currentframe][0] = throw1            # 점수를 딕셔너리의 첫 값으로 입력
        tempresult[currentframe][1] = throw1
        tempresult[currentframe][3] = throw1
        tempresult[currentframe][4] = tempresult[currentframe][4] + tempresult[currentframe][3]
        tempresult[currentframe][5] = strike            # 점수를 딕셔너리의 상태 값을 입력
        print("2.", tempresult)

        #이전 값에 더해야할 값이 있나 확인
        pastframes = frame - 2
        for pastframe in range(pastframes,frame):             # 이전 두 프레임의 상태값을 보고 상태값이 남아있다면 현재값을 이전 프레임의 누적값에 더함
            if pastframe < 0:                                # 0보다 프레임이 작다면
                continue
            else:
                pastkeys = str(pastframe+1) +"프레임"
                if tempresult[pastkeys][5] > 0:             # 과거프레임의 상태가 strike나 spare라면
                    tempresult[pastkeys][4] = tempresult[pastkeys][4] + throw1 #현재 투구의 점수값을 더함
                    tempresult[currentframe][3] = tempresult[currentframe][3] + throw1
                    tempresult[currentframe][4] = tempresult[currentframe][4] + throw1  # 현재 값에 보너스 값 추가
                    tempresult[pastkeys][5] = tempresult[pastkeys][5] - 1  #상태 값에서 1을 줄여줌
                    print("3.", tempresult)

    else :                                                  # 스트라이크가 아닐 경우, 노말, 스페어
        tr2range = defaultpin - throw1
        throw2 = randomthrow(tr2range)
        print("2투구 {} 점입니다.".format(throw2))

        if throw2 == tr2range:
            print("스페어\n")
            tempresult[currentframe][0] = throw1
            tempresult[currentframe][1] = throw2
            tempresult[currentframe][3] = throw1 + throw2
            tempresult[currentframe][4] = tempresult[currentframe][4] + tempresult[currentframe][3] #기존 값에 현재 총합을 더함
            tempresult[currentframe][5] = spare
            print("2.",tempresult[currentframe])

            pastframes = frame - 2
            for pastframe in range(pastframes,frame):
                if pastframe < 0:
                    continue
                else:
                    pastkeys = str(pastframe+1) +"프레임"
                    if tempresult[pastkeys][5] > 0:
                        tempresult[pastkeys][4] = tempresult[pastkeys][4] + throw2
                        tempresult[currentframe][3] = tempresult[currentframe][3] + throw1 + throw2
                        tempresult[currentframe][4] = tempresult[currentframe][4] + throw1 + throw2
                        tempresult[pastkeys][5] = tempresult[pastkeys][5] - 1
                        print("3.", tempresult[currentframe])
        else:
            print("스페어 처리 실패\n")
            tempresult[currentframe][0] = throw1
            tempresult[currentframe][1] = throw2
            tempresult[currentframe][3] = throw1 + throw2
            tempresult[currentframe][4] = tempresult[currentframe][4] + tempresult[currentframe][3]
            tempresult[currentframe][5] = normal
            print("2.",tempresult[currentframe])

            pastframes = frame - 2
            for pastframe in range(pastframes,frame):             # 이전 두 프레임의 상태값을 보고 상태값이 남아있다면 현재값을 이전 프레임의 누적값에 더함
                if pastframe < 0:                                # 0보다 프레임이 작다면
                    continue
                else:
                    pastkeys = str(pastframe+1) +"프레임"
                    if tempresult[pastkeys][5] > 0:             # 과거프레임의 상태가 strike나 spare라면
                        tempresult[pastkeys][4] = tempresult[pastkeys][4] + throw1 #현재 투구의 점수값을 더함
                        tempresult[currentframe][3] = tempresult[currentframe][3] + throw1 + throw2
                        tempresult[currentframe][4] = tempresult[currentframe][4] + throw1 + throw2  # 현재 값에 보너스 값 추가
                        tempresult[pastkeys][5] = tempresult[pastkeys][5] - 1  #상태 값에서 1을 줄여줌
                        print("3.", tempresult[currentframe])


    if frame == 9: #10프레임이면
        print("{}프레임".format(frame+1))
        throw1 = randomthrow(defaultpin)
        print("1투구 {} 점입니다.".format(throw1))
        if throw1 == defaultpin:        # 첫 게임 스트라이크
            throw2 = randomthrow(defaultpin)
            if throw2 == defaultpin:    # 2 게임 스트라이크
                throw3 = randomthrow(defaultpin) # 3 게임 스트라이크

                tempresult["10프레임"][0] = throw1
                tempresult["10프레임"][1] = throw2
                tempresult["10프레임"][2] = throw3
                tempresult["10프레임"][3] = throw1 + throw2 +throw3
                tempresult["10프레임"][4] = tempresult[currentframe][4] + tempresult[currentframe][3]
                tempresult["10프레임"][5] = strike
            else:
                tr3range = defaultpin - throw2 # 범위는 10 - 2번째 투구 점수값
                throw3 = randomthrow(tr3range) # 보너스점수
                throw3 == tr3range
                tempresult["10프레임"][0] = throw1
                tempresult["10프레임"][1] = throw2
                tempresult["10프레임"][2] = throw3
                tempresult["10프레임"][3] = throw1 + throw2 +throw3
                tempresult["10프레임"][4] = tempresult[currentframe][4] + tempresult[currentframe][3]
                tempresult["10프레임"][5] = normal

        else:
            tr2range = defaultpin - throw1 #범위는 10 -1번째 투구 점수값
            throw2 = randomthrow(tr2range) # 2게임 점수

            if throw2 == tr2range:          #스페어 처리
                print("스페어\n")
                throw3 = randomthrow(defaultpin) # 보너스 점수
                tempresult["10프레임"][0] = throw1
                tempresult["10프레임"][1] = throw2
                tempresult["10프레임"][2] = throw3
                tempresult["10프레임"][3] = throw1 + throw2 +throw3
                tempresult["10프레임"][4] = tempresult[currentframe][4] + tempresult[currentframe][3]
                tempresult["10프레임"][5] = spare

            else:
                tempresult["10프레임"][0] = throw1
                tempresult["10프레임"][1] = throw2
                tempresult["10프레임"][2] = 0
                tempresult["10프레임"][3] = throw1 + throw2
                tempresult["10프레임"][4] = tempresult[currentframe][4] + tempresult[currentframe][3]
                tempresult["10프레임"][5] = normal

```
