```python
import random
```

```python
result = [None,None,10] # 프레임,투구수,점수
p_list =[]
point = [None,None,None,None]
s_list =[]
```

```python
# 1에서 9프레임까지의 점수 산출
print("안녕하세요. 이제 볼링 게임을 시작합니다.")

for frame in range(9):
    print("{} 번째 프레임 입니다.".format(frame+1))
    first = random.randint(0,10)  # 첫투구 점수 저장
    print("{} 점입니다.".format(first)) # 점수 출력

    if first == 10:                                                             # 첫 번째 투구가 스트라이크
        print("스트라이크 입니다.\n")
        first = random.randint(0,10)  # 첫투구 점수 저장
        result = [10,0,0]

    else:
        srange = 10 - first                                                                       # 첫 번째 투구가 스트라이크가 아닐 때
        second = random.randint(0,srange)
        print("{} 점입니다.\n".format(second)) # 두번째 점수 출력
        if second+first == 10:                                                  # 두 번째 투구가 스페어처리 했을 때
            print("스페어 입니다.\n")
            result = [first,second,10]
        else :
            result = [first,second,0]                                                            # 두 번째도 스페어 처리 못 했을 때
    p_list.append(result)

```

```python
# 10번째 프레임 (보너스 포함)
print("10 번째 프레임 입니다.")
first = random.randint(0, 10)
print("{} 점입니다.".format(first))

if first == 10:
    print("스트라이크 입니다.")
    second = random.randint(0, 10)
    print("보너스 1구: {} 점입니다.".format(second))
    if second == 10:
        third = random.randint(0, 10)
    else:
        third = random.randint(0, 10 - second)
    print("보너스 2구: {} 점입니다.\n".format(third))
    result = [first, second, third]

else:
    srange = 10 - first
    second = random.randint(0, srange)
    print("{} 점입니다.".format(second))
    if first + second == 10:
        print("스페어 입니다.")
        third = random.randint(0, 10)
        print("보너스 1구: {} 점입니다.\n".format(third))
        result = [first, second, third]
    else:
        print("오픈 프레임입니다.\n")
        result = [first, second, 0]

p_list.append(result)
```

```python
print("\n[프레임별 기록]", p_list)
print("\n[점수 계산]")
```

```python
# 점수 계산

for game in range(10):
    if game == 9:  # 마지막 프레임
        point = [10, p_list[9][0], p_list[9][1], p_list[9][2]]
        print("프레임 10의 점수는 {}입니다.".format(p_list[9][0] + p_list[9][1] + p_list[9][2]))
        s_list.append(point)
    elif p_list[game][0] == 10:  # 스트라이크
        if p_list[game+1][0] == 10:
            if game+2 < 10:
                bonus = p_list[game+2][0]
            else:
                bonus = p_list[game+1][1]  # 마지막 프레임의 2구
            point = [game+1, 10, 10, bonus]
            print("프레임 {}의 점수는 {}입니다.".format(game+1, 10 + 10 + bonus))
            s_list.append(point)
        else:
            point = [game+1, 10, p_list[game+1][0], p_list[game+1][1]]
            print("프레임 {}의 점수는 {}입니다.".format(game+1, 10 + p_list[game+1][0] + p_list[game+1][1]))
            s_list.append(point)
    elif p_list[game][2] == 10:  # 스페어
        point = [game+1, 10, p_list[game+1][0], 0]
        print("프레임 {}의 점수는 {}입니다.".format(game+1, 10 + p_list[game+1][0]))
        s_list.append(point)
    else:
        point = [game+1, p_list[game][0], p_list[game][1], 0]
        print("프레임 {}의 점수는 {}입니다.".format(game+1, p_list[game][0] + p_list[game][1]))
        s_list.append(point)
```

```python
print("\n[최종 점수 리스트]", s_list)
```

```python








# 출력된 점수들로 프레임 점수 산출
for game in range(10):
    if p_list[game][0] == 10 : # 첫 투구로 스트라이크 일때
        if p_list[game+1][0] == 10 : # 두 번째 투구 스트라이크 일때
            point = [game+1,10,10,p_list[game+2][0]]
            print("프레임 {}의 점수는 {}입니다.".format(game+1,10+10+p_list[game+2][0])) # 10 + 10 + 다다음의 첫 투구점수
            s_list.append(point)
        else :
            point = [game+1,10,p_list[game+1][0],p_list[game+1][1]]
            print("프레임 {}의 점수는 {}입니다.".format(game+1,10+p_list[game+1][0]+p_list[game+1][1])) #10 + 다음의 첫투구 + 다음의 두번째 투구
            s_list.append(point)

    elif p_list[game][2] == 10:    # 두 번째로 스페어 처리했을 때
        point = [game+1,10,p_list[game+1][0],0] # 다음 프레임의 첫 타구까지 점수
        print("프레임 {}의 점수는 {}입니다.".format(game+1,10+p_list[game+1][0]))
        s_list.append(point)

    else:  # 스페어 처리 못함
        point = [game+1,p_list[game][0],p_list[game][1],0]
        print("프레임 {}의 점수는 {}입니다.".format(game+1,p_list[game][0]+p_list[game][1]))
        s_list.append(point)

print(s_list)

```
