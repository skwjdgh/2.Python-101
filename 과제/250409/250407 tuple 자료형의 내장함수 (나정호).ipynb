**1. Tuple 자료형의 내장 함수들**

| 함수명     | 설명 |
| ---------- | ---------------------------------------- |
| len(x)     | 해당 튜플의 길이를 반환 |
| max(x)     | 해당 튜플 자료형에서 가장 큰 값 반환 - 비교할 수 없는 경우, 사용 불가능 |
| min(x)     | 해당 튜플에서 가장 작은 값 반환 |
| sum(x)     | 튜플 내부의 값을 합한 값을 반환 |
| sorted(x)  | 해당 튜플의 아이템을 정렬하고 리스트로 반환 |
| any(x)     | 하나라도 참이면 True 반환 |
| all(x)     | 모두 참이면 True 반환 |
| tuple(x)   | 다른 자료형을 튜플로 변환 |

tuple은 불변(immutable) → shuffle() 사용 불가
choice(), sample() 등은 사용 가능 (읽기 전용이라 가능)

```python
#len(x) 해당 튜플의 길이를 반환
print(a, len(a))
print(b, len(b))
print(c, len(c))
print(d, len(d))
print(e, len(e))
print(f, len(f))
```

```python
#max(x) 해당 튜플 자료형에서 가장 큰 값 반환
# print(a, max(a)) 비어있는 공간에서 사용 불가
print(b, max(b))
# print(c, max(c)) 비교를 할 수 없는 경우, 사용 불가
# print(d, max(d)) none 타입 지원하지 않음
# print(e, max(e))
print(f, max(f)) #유니코드 기반으로 가장 큰 값을 반환
```

```python
#min(x) 해당 튜플에서 가장 작은 값 반환
# print(a, min(a)) 비어있는 공간에서 사용 불가
print(b, min(b))
# print(c, min(c)) 비교를 할 수 없는 경우, 사용 불가
# print(d, min(d)) none 타입 지원하지 않음
print(e, min(e)) #()이 존재하기 때문에 사용 가능
print(f, min(f)) #유니코드 기반으로 가장 큰 값을 반환
# 튜플끼리는 비교 가능
temp = [(1,2,3),(4,5,6)]
min(temp)
```

```python
#sum(x) # 튜플 내부의 값을 합한 값을 반환
print(sum(a)) # 공백에는 사용 가능
print(sum(b))
# print(sum(c)) 중첩 튜플에서는 사용 불가
# print(sum(f)) 문자들의 합은 불가능하다.
# print(sum(d)) none 값의 계산 불가능
```

```python
#sorted(x) 해당 튜플의 아이템을 정렬하고 리스트로 반환
print(sorted(a))
print(type(sorted(a))) # 리스트 형으로 반환됨!
print(sorted(b))
# print(sorted(c)) 중첩은 정렬 불가
# print(sorted(d)) non 데이터도 정렬 불가
print(sorted(f)) # 유니코드 기준
print(sorted(h)) # key 값 기준으로
```

```python
#any(x) 	하나라도 참이면 True 반환
print(any(a)) # 값이 없으므로
print(any(b))
print(any(c))
print(any(d)) # none 자료형이 있으므로
print(any(e))
```

```python
#all(x) 모두 참이면 True 반환
print(all(a)) # 비어있는 경우 항상 참을 반환함
print(all(b))
print(all(c))
print(all(d)) # none 자료형이 있으므로
print(all(e)) # none 자료형이 있으므로
```

```python
#tuple(x)
print(f)
print(g)
print(h)
print(i) # 내부 아이템의 형변환은 없다.
```

**2. Tuple 자료형의 메서드**

| 메서드명                 | 설명 |
| ------------------------ | ---------------------------------------- |
| count(x)                 | 입력한 값을 튜플 내에서 찾아 그 개수를 반환함 |
| index(x[, start[, end]]) | 범위가 정해져 있다면 범위 내에서, 아니라면 해당 튜플 내에서 찾는 값의 인덱스 값을 반환한다. |

```python
# count(x) 입력한 값을 튜플 내에서 찾아 그 개수를 반환함

print(a, a.count(1))
print(b, b.count(1))
print(c, c.count(1)) # 중첩된 아이템들의 인자는 인식 못함
print(d, d.count(1))
print(e, e.count(1))
```

```python
#index(x[, start[, end]]) 범위가 정해져 있다면 범위 내에서, 아니라면 해당 튜플 내에서 찾는 값의 인덱스 값을 반환한다.
# tuple.index(x, start)
# tuple.index(x, start, end)
# print(a, a.index()) # 데이터 없음, 최소 1개의 아이템 필요
print(b, b.index(3)) # 찾는 값의 리스트 내에서의 위치를 알려줌
# print(c, c.index(2)) # 찾는 값이 튜플 내에 없음
# print(d, d.index(2))
# print(e, e.index(2))
print(b, b.index(2,1,2)) # 범위를 정한 경우에도 리스트 내에서 위치를 알려주지 범위 내 위치값을 반환하는 것은 아님
```
