### 1. list 자료형의 내장 함수들

| 함수명     | 설명 |
| ---------- | ---------------------------------------- |
| len(x)     | 리스트의 길이 값을 반환 |
| max(x)     | 리스트내 가장 큰 값을 반환 |
| min(x)     | 리스트에서 가장 작은 값을 반환 |
| sum(x)     | 연산이 가능한 리스트의 합 계산(문자는 계산 불가능) |
| sorted(x)  | 리스트를 정렬하여 반환, 원본 리스트는 변경되지 않음 |
| any(x)     | 하나라도 참이면 True 반환 |
| all(x)     | 모두 참이면 True 반환 |
| list(x)    | 다른 자료형을 리스트로 변환 |

함수는 파라미터로 특정 작업을 수행, 메서드는 클래스로 특정 작업을 수행

```python
#len(x) 리스트의 길이 값을 반환
a1 = len(a)
a2 = len(b)
a3 = len(c)
a4 = len(d)

print(a1,a2,a3,a4)
```

```python
#max(x) 리스트내 가장 큰 값을 반환
#max(a) 빈공간 사용 불가
max(b)
#max(c) 리스트와 None 사용 불가
#max(d)
#max(e)
print(max(b))
```

```python
#min(x) 리스트에서 가장 작은 값을 반환
#min(a)
min(b)
#min(c)
#min(d)
f= [b,c,d]
#print(min(f))
min(e)
print(min(b),min(e)) # min(e)는 리스트의 가장 크기가 작은 값이 들어감, min(e)는 연산가능 min(f)는 연산불가 => 가장 앞에서 비교할 것이 없어 최소값이 됨
```

```python
#sum(x) #연산이 가능한 리스트의 합 계산(문자는 계산 불가능)
print(sum(a))
print(sum(b))
#sum(c)
#sum(d)
#sum(e)
#sum(f) #문자들은 계산 불가
```

```python
#sorted(x) 리스트를 정렬하여 반환, 원본 리스트는 변경되지 않음
print(sorted(a)) #비어있는 리스트형도 사용 가능
print(sorted(b))
#print(sorted(c))
#print(sorted(d)) none 사용불가능
#print(sorted(e))
print(f, sorted(f)) # 유니코드 기준으로 정렬됨
```

```python
#any(x) 하나라도 참이면 True 반환
print(any(a))  #리스트 내 아이템 없음
print(any(b))
print(any(c))
print(any(d))  #리스트 내 None 선언
print(any(e))
print(any(f))
```

```python
#all(x) 모두 참이면 True 반환
print(all(a)) # 내부는 비어있는데 왜 참인가? 빈 자료형들은 항상 참반환 () {} []
print(all(b))
print(all(c))
print(all(d)) # None으로 차서?
print(all(e)) # all(x) 는 내부 요소들을 검사하는 것임, all(a)일때는 참이지만 f= [all(a)]일 때, all(f) 는 거짓이됨
print(all(f))
```

```python
#list(x) 다른 자료형을 리스트로 변환
print(list(a))
print(list(strt)) # 문자열을 잘라서 리스트를 만듬
print(list(tuplet)) # 선언되었지만 자료형 자체가 변경됨
print(list(dictt)) # 딕셔너리는 key 값만 리스트가 됨
print(list(dictt.values())) # 딕셔너리 value들만 꺼냄
```

**2. 리스트 자료형의 메서드**

| 메서드명                 | 설명 |
| ------------------------ | ---------------------------------------- |
| append(x)                | 리스트 끝에 항목을 추가 |
| extend(iterable)         | 반복 가능한 객체들을 리스트의 끝에 추가 |
| insert(i, x)             | i번째 위치에 x 삽입 |
| remove(x)                | 리스트 내 x 제거 (없으면 에러 발생) |
| pop([i])                 | i번째 리스트을 제거 후 반환, 매개변수가 없을 때, 뒤부터 제거함. |
| clear()                  | 리스트 내의 x 제거 ( 값이 없으면 에러 발생) |
| index(x[, start[, end]]) | 리스트 내 x의 인덱스 반환 ( 값이 없으면 에러 ) |
| count(x)                 | 리스트에서 x값의 개수를 찾음 |
| sort()                   | 리스트 정렬(기본은 내림차순 - 논리값을 통해 오름차순 변경 가능) |
| reverse()                | 아이템들의 순서를 앞뒤로 바뀜, 아이템의 내부 자료는 변화 없음 |
| copy()                   | 리스트의 얕은 복사본 반환 |

```python
#append(x) 리스트 끝에 항목을 추가
b.append(25) # 리스트 뒤로 자료 추가
print(b)
c.append(25) # 리스트 아이템이 아닌 리스트에 붙음
print(c)
#strt.append(25) 동일 자료형 에서만 출력 가능
#tuplet.append(25)
#dictt.append(25)
```

```python
#extend(iterable) 반복 가능한 객체들을 리스트의 끝에 추가
g=[1,2,3]
# print(g.extend(a)) none 반환
g.extend(a)
print(g)
g.extend(b)
print(g)
g.extend(c)
print(g)
g.extend(d)
print(g)
g.extend(e) # 빈 리스트의 경우, 요소로서의 리스트는 리스트에 추가되지만 단독 아이템으로서의 리스트는 사라져버림
print(g)
g.extend(strt)
print(g)
g.extend(tuplet)
print(g)
g.extend(dictt) #같은 리스트를 제외한 자료형은 자료형을 해체하여 리스트에 추가됨
print(g)
```

```python
#insert(i, x) i번째 위치에 x 삽입
print(a, a.insert(3,3)) #list.insert(i,x) 는 반환값이 없음! 빈 공간에는 위치 상관없이 입력가능
print(b, b.insert(3,3))
print(c, c.insert(3,3))
print(d, d.insert(3,3))
print(e, e.insert(3,3))
```

```python
#remove(x) 리스트 내의 x 제거 ( 값이 없으면 에러 발생)
# print(a.remove(3)) none 반환
print(a, a.remove(4))
print(b, b.remove(4))
print(c, c.remove(4))
print(d, d.remove(4))
print(e, e.remove(4))
```

```python
#pop([i]) i번째 리스트 아이템을 제거 후 반환, 매개변수가 없을 때, 뒤부터 제거함.
print(a, a.pop(3)) #빈 리스트에서는 오류 발생
print(b, b.pop(3)) # pop 은 반환값이 존재, 나온 원본에서는 해당 항목이 제거됨
print(c, c.pop(1)) # 자료형 아이템도 문제없이 꺼냄
print(d, d.pop(1))   # None 도 문제 없이 꺼냄
print(e, e.pop(1))
```

```python
#clear() 리스트의 모든 항목을 제거
print(a, a.clear())
print(b, b.clear())
print(c, c.clear())
print(d, d.clear())
print(e, e.clear())
```

```python
# index(x[, start[, end]]) 리스트 내 x의 인덱스 반환 ( 값이 없으면 에러 )
# list.index(x, start)
# list.index(x, start, end)
# print(a, a.index()) 데이터 없음
print(b, b.index(3)) # 찾는 값의 리스트 내에서의 위치를 알려줌
# print(c, c.index(3))
# print(d, d.index(3))
# print(e, e.index(3))
print(b, b.index(2,1,2)) # 범위를 정한 경우에도 리스트 내에서 위치를 알려주지 범위 내 위치값을 반환하는 것은 아님
```

```python
#count(x) 리스트에서 x값의 개수를 찾음

# print(a, a.count())
print(a, a.count(1))
print(b, b.count(1)) #다른 자료형 내의 아이템은 찾지 않음
print(c, c.count(1))
print(d, d.count(1))
print(e, e.count(1))
```

```python
#sort(key=None, reverse=False) 리스트 정렬(기본은 내림차순 - 논리값을 통해 오름차순 변경 가능)
print(a, a.sort())
b=[4,3,1,2]
print(b, b.sort(reverse=True)) # 내림차순
print(b, b.sort())

temp = [[1,2,3,4],[4,5,6,7]]
[1,2,3,4] < [4,5,6,7] #요소끼리 비교가 가능할 때!!!, 비교식을 사용할 수 있음
print(temp, temp.sort(reverse=True))
# [[1,2,3,4],[4,5,6,7]] < [1,2,3,4] #리스트 묶음 사용 불가
# [[[1,2,3,4],[4,5,6,7]],[1,2,3,4]] < [1,2,3,4,5]
# print(c, c.sort(reverse=False)) # list 자료형이라 순서를 정할 수 없음, 리스트 간에 순서를 정해둔다면 사용할 수 있음
# print(d, d.sort(reverse=False)) none 타입을 지원하지 않음
# print(e, e.sort(False))
comb = a + f
print(comb)
print(comb.sort()) #유니코드 순서에 따라 정렬
print(comb)
```

```python
#reverse() 아이템들의 순서를 앞뒤로 바뀜, 아이템의 내부 자료는 변화 없음, 정렬 시 none 반환, 데이터가 없어도 정상동작
print(a, a.reverse())
print(b, b.reverse())
print(c, c.reverse())
print(d, d.reverse())
print(e, e.reverse())
```
