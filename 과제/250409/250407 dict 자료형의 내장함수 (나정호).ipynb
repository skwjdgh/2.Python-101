### 1.dict 자료형의 함수들

| 고정함수명    | 설명 |
| ------------- | ---------------------------------------- |
| len(x)        | 딕셔너리 키밸류쌍 개수 반환 |
| dict()        | 딕셔너리 생성 및 다른 자료형의 변환 |
| type(x)       | 객체의 자료형 반환 |
| str(x)        | 딕셔너리 자료형을 문자열로 반환 |
| sorted(x)     | 키를 정렬하여 리스트로 반환 |
| all(x)        | 모든 키가 참이면 True 반환 |
| any(x)        | 하나라도 참인 키가 있으면 True 반환 |
| max(x)        | 가장 큰 키 반환 |
| min(x)        | 가장 작은 키 반환 |
| sum(x)        | 키의 합 반환 |


```python
#len(x) 딕셔너리 키밸류쌍 개수 반환
print(len(a))
print(len(b))
print(len(c)) # 중첩문의 자료는 계산하지 않음
```

```python
# 중첩 Dictionary 자료형들
l1 = [[1, "리"], [2, "스"],[3, "트"]]
l2 = [[1, "동"], [1, "일"],[1, "키"]]
t1 = ((1, "튜"), (2, "플"),(3, "형"))
t2 = ([1, "리"], [2, "튜"],[3, "형"])
t3 = ((1, "동"), (1, "일"),(1, "키"))
# s1 = {{1, "셋"},{2, "자"},{3, "료"}}
s2 = {(1, "튜"),(2, "플"),(3, "셋")}
# s3 = {[1, "리"],[2, "스"],[3, "셋"]}
s4 = {(1, "동"),(1, "일"),(1, "키")}
```

```python
#dict() 딕셔너리 생성 및 다른 자료형의 변환
# 다른 자료형 초기화
# strt = "문자열을 테스트하자"
# tuplet = (1,2,3,4,5,6)
# listt = [1,2,3,4,5,6]
# f= dict(strt)
# g= dict(tuplet)
# h= dict(listt)
# print(f)
# print(g)
# print(h)
print(dict(l1))
print(dict(l2)) # 마지막 값이 키와 쌍이 됨
print(dict(t1))
print(dict(t2))
print(dict(t3)) # 마지막 값이 키와 쌍이 됨
print(dict(s2))
print(dict(s4)) # set 자료형의 경우, 순서가 랜덤이기 때문에 어떤 값이 나올지 모름 !!
```

```python
#type(x) 객체의 자료형 반환
print(type(a))
print(type(b))
```

```python
#str(x)  딕셔너리 자료형을 문자열로 반환
print(a, str(a))
print(b, str(b))
print(c, str(c))
print(type(str(b)))

for rangeb in range( len(str(b)) ):
    print(str(b)[rangeb])
# 키밸류의 모든 아이템들이 문자열로 바뀜
```

```python
#sorted(x)  키를 정렬하여 리스트로 반환
a = {3: '들',2: '자',1: '문'} # 순서가 생김
print(a, sorted(a))
print(b, sorted(b))
print(c, sorted(c))
```

```python
#all(x) 모든 키가 참이면 True 반환
print(a, all(a)) #빈 공간은 항상 True
print(b, all(b))
print(c, all(c))
a = {None:2, 2:'자',1: '문'}
print(a, all(a)) #None으로 인해 false
b = {3:None, 2:'자',1: '문'}
print(b, all(b)) #key가 none 일때는 false, value가 none 일때는 true
```

```python
#any(x)
print(a, any(a)) # all 과 다르게 False
print(b, any(b))
print(c, any(c))
a = {None:2, 2:'자',1: '문'}
b = {3:None, 2:'자',1: '문'}
print(a, all(a)) # 역시 key가 none 일때 false 임
print(b, all(b))
```

```python
#max(x)
# print(a, max(a))
print(b, max(b))
print(c, max(c))
a = {"사과":2, "포도":'자',"감자": '문'}
print(a, max(a)) # 유니코드 기준 결과값임
```

```python
#min(x)
# print(a, min(a))
print(b, min(b))
print(c, min(c))
a = {"사과":2, "포도":'자',"감자": '문'}
print(a, min(a))
```

### 메서드 (메서드: 딕셔너리에서 직접 호출)

| 메서드명        | 설명 |
|-----------------|----------------------------------------------------|
| clear           | 모든 항목 제거 |
| copy            | 얕은 복사본 반환 |
| fromkeys        | 순서가 있는 시퀀스 들로 새로운 딕셔너리 생성, 시퀸스의 값들이 key값이 됨 |
| get             | 특정키 반환, 키가 없으면 기본값 반환 |
| items           | (키, 값) 쌍 목록 반환 |
| keys            | 키 목록 반환 |
| pop             | 특정키 제거 |
| popitem         | 딕셔너리 뒤에서 부터 반환하며 제거함 |
| setdefault      | 입력 값을 키로 하여 value 값을 반환, 키가 없으면 값 설정 후 반환 |
| update          | 다른 key value 쌍을 통해 내용 변경 |
| values          | value 목록 반환 |

```python
#clear()
print(a, a.clear())
print(b, b.clear())
print(b, b.clear())
```

```python
#copy() 얕은 복사본 반환
print(a,a.copy()) #메모리 주소가 다른 dictionary 반환(Shallow Copy)
print(id(a))
print(id(a.copy()))
```

```python
#fromkeys(seq[, value]) 순서가 있는 시퀀스 들로 새로운 딕셔너리 생성, 시퀸스의 값들이 key값이 됨
k = [1,2,3,4,5,6,7]
l = {1:5, 2:7, 3:8}
m = ["a","b","c","d"]

print(a, a.fromkeys(k))
print(a, a.fromkeys(l)) #키 에 맞추어 기존 값이 대입?
print(b, b.fromkeys(k))
print(b, b.fromkeys(m))
```

```python
#get(key[, default]) 특정키 반환, 키가 없으면 기본값 반환
print(a, a.get(1))
print(b, b.get(1))
print(c, c.get(1))
```

```python
#items() key:value 목록 반환
print(a,a.items())
"""dictionary 자료형은 굉장히 특이한 뷰 객체를 반환하도록 되어있음. dict_items([]) 처럼 dict_keys, dict_values 등 다양한 메서드에서 반환하는 클래스를 나타냄"""
print(b,b.items())
print(c,c.items())
```

```python
#keys() key 값 반환
print(a, a.keys())
print(b, b.keys())
print(c, c.keys())
```

```python
#pop(key[, default]) 특정키 제거
#print(a, a.pop(1))
print(b, b.pop(1)) # 꺼내서 원래 dict에서는 제외함
print(c, c.pop(1))
```

```python
#popitem() 딕셔너리 뒤에서 부터 반환하며 제거함
#print(a, a.popitem()) # 비어 있는 경우 사용 불가
print(b, b.popitem())
print(c, c.popitem())
```

```python
#setdefault(key[, default])
print(a, a.setdefault(1)) #값이 없으면 키값을 만들어 none value를 반환하게됨
print(b, b.setdefault(1))
print(c, c.setdefault(1))
```

```python
#update([other]) 다른 key value 쌍을 통해 내용 변경
print(a, a.update(b))
print(a, a.update(dict(l1)))
# print(a, a.update([1,2]))
print(a, a.update({1:2})) # 같은 키의 경우, 교체
print(a, a.update({4:2})) # 다른 키의 경우, 추가
# 다른 딕셔너리 혹은 키 밸류값으로만 사용 가능
```

```python
#values() #value 목록 반환
print(a, a.values())
print(b, b.values())
print(c, c.values())
```
