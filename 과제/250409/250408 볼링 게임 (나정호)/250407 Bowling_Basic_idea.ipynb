```python
import random                                                                    # 랜덤 함수 사용을 위한 모듈 호출
```

```python
# 필요 전역변수 정리

result = [None,None,0]                                                          # 1투구, 2투구, 3투구(혹은 스페어 여부)
p_list =[]                                                                      # 결과들을 저장할 리스트
point = [None,None,None]                                                        # 스트라이크 및 스페어 처리를 위한 리스트 프레임, 1투구, 2투구, 3투구
s_list =[]                                                                      # 점수들을 저장할 리스트
```

```python
# 1 ~ 9프레임 게임 플레이

print("안녕하세요. 이제 볼링 게임을 시작합니다.")

for frame in range(9):                                                          # 1 ~ 9회까지의 프레임 반복
    print("{} 번째 프레임 입니다.".format(frame+1))

    first = random.randint(0,10)                                                # 첫번째 투구 점수 생성
    print("첫 번째 투구 : {} 점입니다.".format(first))

    if first == 10:                                                             # 조건문 1-1 : 첫 번째 투구가 스트라이크
        print("스트라이크 입니다.\n")
        first = random.randint(0,10)
        result = [10,0,0]                                                       # 스트라이크 결과를 result 리스트에 저장

    else:                                                                       # 조건문 1-2 : 스트라이크가 아닌 모든 경우
        srange = 10 - first                                                     # 첫 투구 결과를 제외한 새로운 범위 생성
        second = random.randint(0,srange)                                       # 새로운 범위 기준으로 두번째 투구 점수 생성
        print("두 번째 투구 : {} 점입니다.".format(second))

        if second+first == 10:                                                  # 조건문 2-1 : 두 번째 투구가 스페어 처리됨
            print("스페어 입니다.\n")
            result = [first,second,10]                                          # 스페어 결과를 result 리스트에 저장, 리스트 세번째 위치에 10으로 표기
        else :                                                                  # 조건문 2-2 : 두 점수의 합이 10점이 되지 않음
            result = [first,second,0]                                           # 스트라이크와 스페어 없이 result 리스트에 저장
    p_list.append(result)                                                       # 1 ~  9회까지의 result 리스트들을 p_list에 저장
```

```python
# 10 프레임 게임 플레이

print("10 번째 프레임 입니다.")

first = random.randint(0, 10)                                                   # 10프레임의 첫 번째 투구 점수 생성

print("{} 점입니다.".format(first))

if first == 10:                                                                 # 조건문 1-1 : 첫 번째 투구가 스트라이크
    print("스트라이크 입니다.\n")
    second = random.randint(0, 10)                                              # 스트라이크일 경우, 새로운 범위는 0~10 까지임
    print("보너스 투구: {} 점입니다.".format(second))
    if second == 10:                                                            # 조건문 2-1 : 두 번째 투구가 스트라이크
        third = random.randint(0, 10)                                           # 스트라이크일 경우, 새로운 범위 역시 0~10 까지임
    else:
        third = random.randint(0, 10 - second)                                  # 조건문 2-2 : 스트라이크가 아님
    print("보너스 투구: {} 점입니다.\n".format(third))                           # 스트라이크가 아닐 경우, 새로운 범위는 10- second 가 됨
    result = [first, second, third]                                             # 결과를 result 리스트에 저장함

else:                                                                           # 조건문 1-2 : 첫 번째 투구가 스트라이크가 아님
    srange = 10 - first                                                         # 스트라이크일 경우, 새로운 범위는 10-first 임
    second = random.randint(0, srange)
    print("{} 점입니다.".format(second))
    if first + second == 10:                                                    # 조건문 2-1 : 두 번째 투구로 스페어 처리 여부
        print("스페어 입니다.\n")
        third = random.randint(0, 10)                                           # 스페어 처리시, 새로운 범위는 0~10 까지임
        print("보너스 투구: {} 점입니다.\n".format(third))
        result = [first, second, third]                                         # 결과를 result 리스트에 저장함
    else:                                                                       # 조건문 2-2 : 스페어 처리를 못해 보너스 경기 없음
        print("경기 종료입니다.\n")
        result = [first, second, 0]                                             # 결과를 result 리스트에 저장함
p_list.append(result)
```

```python
# 점수 계산

for game in range(10):                                                          # 1 ~ 10 프레임 까지의 점수 계산
    if game == 9:  # 마지막 프레임                                              # 조건문 1-1 : 10프레임에서의 점수 계산
        point = p_list[9][0]+ p_list[9][1]+ p_list[9][2]
        print("프레임 10의 점수는 {}입니다.".format(point))  # 10 프레임의 점수
        s_list.append(point)                                                    # 새로운 리스트를 s_list에 저장
    elif p_list[game][0] == 10:                                                 # 조건문 1-2 : 1~9 프레임까지의 결과 중, 스트라이크일 경우
        if p_list[game+1][0] == 10:                                             # 조건문 2-1 : 두번째 투구에서 프레임에서 스트라이크가 발생했을 때
            if game+2 < 10:                                                     # 조건문 3-1 : 1~9 프레임까지 점수 계산법 보너스는 다다음 프레임의 첫 투구값
                bonus = p_list[game+2][0]
            else:                                                               # 조건문 3-2 : 10 프레임에서의 점수 계산법 보너스는 다다음 프레임의 2번째 투구값까지
                bonus = p_list[game+1][1]
            point = 10 + 10 + bonus                                             # point 리스트에 1,2 투구가 스트라이크 일때 점수를 저장함
            print("프레임 {}의 점수는 {}입니다.".format(game+1, point))
            s_list.append(point)                                                # 기록된 점수를  s_list에 저장함
        else:                                                                   # 조건문 2-2 : 두번째 투구가 스트라이크가 아닐 때
            point = 10 + p_list[game+1][0] + p_list[game+1][1]                  # 점수는 10 + 다음게임의 첫 점수와 다음게임의 두번째 점수
            print("프레임 {}의 점수는 {}입니다.".format(game+1, point))
            s_list.append(point)
    elif p_list[game][2] == 10 and game < 9:                                    # 조건문 1-3 : 9프레임까지 두 번째 기록이 스페어라면 'result의 마지막 아이템으로 스퀘어 표기'
        point = 10 + p_list[game+1][0]                                          # 점수는 10점 + 다음 프레임의 첫 투구 점수
        print("프레임 {}의 점수는 {}입니다.".format(game+1, point))
        s_list.append(point)
    elif  p_list[game][2] != 10:                                                # 조건문 1-4 : 9프레임까지 두 번째 투구로 스페어처리를 못했다면
        point = p_list[game][0]+ p_list[game][1]                                # 점수는 첫 투구 점수와 두번째 투구 점수
        print("프레임 {}의 점수는 {}입니다.".format(game+1, point))
        s_list.append(point)
```

```python
# 점수 출력
print("\n최종 점수 계산 테이블", s_list)
print("최종 점수는 {}점 입니다.".format(sum(s_list)))
```
