```python
import random
```

### 1. 기본 난수 생성 함수
| 고정함수명                      | 설명 |
|---------------------------------|--------------------------------------------|
| random()                        | 0.0 이상 1.0 미만의 float 실수 반환 |
| uniform(a, b)                   | a 이상 b 이하의 float 실수 반환 |
| randint(a, b)                   | a 이상 b 이하의 정수 반환 (양 끝 포함) |
| randrange([start], stop[, step])| range(start, stop, step) 중 무작위 선택 |

```python
# random() 0과 1 사이의 실수값 반환
a = random.random()
print(a)
```

```python
# uniform(a, b) a와 b 사이의 실수값 반환
a = 1
b = 2
c = random.uniform(a,b)
print(c)
```

```python
# randint(a, b)
b = 100
c = random.randint(a,b)
print(c)
```

```python
# randrange([start], stop[, step]) # randint 와 동일하게 정수값 반환, 하지만 step을 통해 증가 간격 조건을 줄 수 있음
c = random.randrange(a,b,5) #start,stop, step
print(c)
```

### 2. 시퀀스 관련 함수
| 고정함수명 | 설명 |
|-----------|--------------------------------------------|
| choice(seq)                            | 시퀀스에서 임의 항목 하나 반환 |
| choices(population, weights=None, k=1) | 복원 추출 (중복 허용)로 k개 항목 반환 |
| sample(population, k)                  | 비복원 추출 (중복 없음)으로 k개 항목 반환 |
| shuffle(seq)                           | 시퀀스 항목을 제자리에서 무작위 섞기 |

```python
# 시퀸스 관련 변수 초기화
a = "문자열을 확인합시다."
b = [1,2,3,4,5,6]
```

```python
#choice(seq)
result = random.choice(a)
print(result)
result = random.choice(b)
print(result)
```

```python
#choices(population, weights=None, k=1) #weights는 가중치, 일반적으로 리스트, 튜플을 통해 값을 직접 전달 , k는 반환할 항목의 개수를 표현
weight = [1,1,3,1,1,1]
result = random.choices(b,weights=weight,k=7)
print(result)

weight = [3,1,1,1,2,1]
result = random.choices(b,weights=weight,k=3)

print(result)
```

```python
#sample(population, k) k는 원본 자료형보다 클 수 없다.

result = random.sample(b,5)
print(result)
result = random.sample(b,3)
print(result)
```

```python
#shuffle(seq) 문자열에서는 사용 불가능함
print(b)
result = random.shuffle(b)
print(b)
#print(a)
#result = random.shuffle(a)
#print(a)
```

### 3. 시드 및 상태 제어 함수 📌
| 고정함수명 | 설명 |
|-----------|--------------------------------------------|
| seed(a=None, version=2) | 난수 시드 설정 (재현성 확보용) |
| getstate()              | 현재 난수 생성기의 내부 상태 가져오기 |
| setstate(state)         | 난수 생성기의 상태 복원 |

```python
#seed(a=None, version=2) 시드 값을 고정해 결과를 재현할 수 있음

random.seed(2) # 시드 고정
print(random.random())
print(random.randint(1, 10))
random.seed(2)
print(random.random())
print(random.randint(1, 10))
random.seed(2)
print(random.random())
print(random.randint(1, 10))
random.seed(2)
print(random.random())
print(random.randint(1, 10))
```

```python
#getstate() ,#setstate(state)
random.seed(123)
state = random.getstate()

print("초기 값1 :", random.random())
print("초기 값2 :", random.random())
print("초기 값3 :", random.random())

random.setstate(state)

print("복원 값1 :", random.random())
print("복원 값2 :", random.random())
print("복원 값3 :", random.random())
```

### 4. 확률 분포 관련 함수
각 분포에 기반한 난수를 생성하는 함수들

| 고정함수명 | 설명 |
|-----------|--------------------------------------------|
| normalvariate(mu, sigma)  | 정규 분포 기반 난수 생성 |
| gauss(mu, sigma)          | 정규 분포 기반 난수 생성 (가우시안) |
| betavariate(alpha, beta)  | 베타 분포 (alpha, beta) 기반 난수 생성 |
| expovariate(lambd)        | 지수 분포 기반 난수 생성 |
| gammavariate(alpha, beta) | 감마 분포 기반 난수 생성 |
| triangular(low, high, mode) | 삼각 분포 기반 난수 생성 |
| lognormvariate(mu, sigma) | 로그 정규 분포 기반 난수 생성 |
| weibullvariate(alpha, beta) | 와이블 분포 기반 난수 생성 |
| vonmisesvariate(mu, kappa)| 폰 미제스 분포 (원형 정규분포) |
| paretovariate(alpha)      | 파레토 분포 기반 난수 생성 |





```python
mu = 0        # 평균
sigma = 1     # 표준편차
alpha = 2     # 베타 분포 및 감마 분포의 파라미터
beta = 3      # 베타 분포 및 감마 분포의 파라미터
lambd = 1     # 지수 분포의 람다
low = 1       # 삼각분포의 최소값
high = 10     # 삼각분포의 최대값
mode = 5      # 삼각분포의 최빈값
kappa = 1     # 집중도 (kappa 값이 클수록 분포가 좁아짐)
alpha = 2     # 파레토 분포의 형태 파라미터
```

```python
print(random.normalvariate(mu, sigma))      # 정규 분포는 평균(mu)을 중심으로 데이터가 대칭적으로 분포하는 분포
print(random.gauss(mu, sigma))              # 가우시안 분포는 대표적인 정규 분포의 하나로 용도에 따라 구분해 놓음
print(random.betavariate(alpha, beta))      # 베타 분포는 [0, 1] 범위에서 값을 가지며, alpha와 beta는 분포의 모양을 결정하는 파라미터로, 이 값에 따라 분포가 비대칭적 -좌집중, 우집중
print(random.expovariate(lambd))            # 지수 분포는 lambd는 사건이 발생하는 평균 빈도(람다), lambd가 클수록 사건 발생이 더 자주 일어남 - 한쪽으로 긴 꼬리가 있는 형태
print(random.gammavariate(alpha, beta))     # 감마 분포는 지수 분포의 일반화된 형태로, 여러 사건의 발생 간격을 합친 분포 - 한쪽으로 긴 꼬리가 있는 형태
print(random.triangular(low, high, mode))   # 삼각 분포는 최소값(low), 최대값(high), 최빈값(mode)을 정의하여 데이터가 특정 값에 집중되도록 모델링 - 삼각형 모양
print(random.lognormvariate(mu, sigma))     # mu와 sigma는 로그 변환된 정규 분포의 평균과 표준편차  - 오른쪽으로 긴 꼬리
print(random.weibullvariate(alpha, beta))   # - 기울어진 형태, 긴 꼬리
print(random.vonmisesvariate(mu, kappa))    # 폰 미제스 분포는 원형 정규 분포라고도 불리며, 주로 각도 데이터를 모델링, mu는 평균 각도, kappa는 집중도 - 원형으로 대칭적인 종 모양
print(random.paretovariate(alpha))          # 많은 현상에서 일부 값들이 대부분의 결과를 차지하는 분포, alpha는 분포의 형태를 결정하는 파라미터 - 오른쪽으로 긴 꼬리
```
