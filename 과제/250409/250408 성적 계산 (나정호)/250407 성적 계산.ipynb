```python
import random
```

```python
# 학년, 과목, 반의 입력 여부에 따른 평균(average), 최고점(high), 최저점(low)을 구하는 함수

def get_average(ginput,subinput,school,rinput=None):
    cal = []                                                                    # 조건문에 따른 value 값을 모으기 위한 리스트

    if ginput not in school:                                                    # ginput(학년) 값이 school 내에 존재하지 않는다면 재입력 요청
        print("{} 학년은 없습니다. 다시 입력하세요.".format(ginput))
        return

    if subinput not in ["국어","영어","수학"]:                                  # subinput 값(과목)이 국어, 영어, 수학 중에 하나가 아니라면 재입력 요청
        print("{} 과목은 유효하지 않습니다. 다시 입력하세요.".format(subinput))
        return

    if rinput:                                                                  # 만약에 rinput 값(반)이 존재하면 아래 조건문 실행
        if rinput not in school[ginput]:                                        # rinput 값(반)이 school[ginput](학년)에 존재하지 않는다면 재입력 요청
            print("{} 은 없습니다. 다시 입력하세요.".format(rinput))
            return

        cal = [student[subinput] for student in school[ginput][rinput].values()]# rinput 값이 존재한다면 cal 값은 school[ginput][rinput]의 value 들을 student에 할당하고 이를 과목으로 추출해서 cal에 리스트로 저장한다.

        average = sum(cal)/len(cal)
        high = max(cal)
        low = min(cal)
        print("{}의 {} {} 과목 평균 점수는 {}점 입니다." .format(ginput,rinput,subinput,average))
        print("{}의 {} {} 과목 최고점은 {}, 최저점은 {}." .format(ginput,rinput,subinput,high,low))
        exit()                                                                  # 조건문 종료

    else :                                                                      # 만일 rinput이 존재하지 않는다면
        for room in school[ginput].values():                                    # 딕셔너리 school 자료형에 ginput라는 key를 가지는  value 들을 room에 저장함
            for student in room.values():                                       # 딕셔너리 room 자료형에 있는 value들을 student에 저장함
                if subinput in student:                                         # 조건문 : 만약 subinput이 student에 존재한다면 subinput에 따른 value 값들은 cal에 저장함
                    cal.append(student[subinput])

        average = sum(cal)/len(cal)
        high = max(cal)
        low = min(cal)
        print("{}의 {} 과목 평균 점수는 {}점 입니다." .format(ginput,subinput,average))
        print("{}의 {} 과목 최고점은 {}, 최저점은 {}." .format(ginput,subinput,high,low))
        exit()                                                                  # 조건문 종료
    return
```

```python
# 모든 학생들의 과목별 평균을 구하는 함수

def get_subaverage(school):
    lcal=[]
    ecal=[]
    mcal=[]

    for grade in school.values():
        for room in grade.values():
            for score in room.values():
                lcal.append(score["국어"])
                ecal.append(score["영어"])
                mcal.append(score["수학"])

    laverage = sum(lcal)/len(lcal)
    eaverage = sum(ecal)/len(ecal)
    maverage = sum(mcal)/len(mcal)

    print("국어 과목의 전체 학생들의 평균은 {}입니다.".format(laverage))
    print("영어 과목의 전체 학생들의 평균은 {}입니다.".format(eaverage))
    print("수학 과목의 전체 학생들의 평균은 {}입니다.".format(maverage))
```

```python
# 학교의 학생들의 성적을 딕셔너리 자료형으로 생성

school = {}                                                                     # 학년 데이터를 저장할 school 초기화, 반복문을 돌며 학년 데이터를 저장
for grade in range(6):
    room = {}                                                                   # 반 데이터를 저장할  room 초기화, 반복문을 돌며 반 데이터를 저장
    for room_num in range(5):
        student = {}                                                            # 학생 데이터를 저장할 student 초기화, 반복문을 돌며 학생 데이터를 저장
        for stu_num in range(10):
            point = {                                                           # 학생의 국어, 영어, 수학 점수 생성
                '국어': random.randint(0, 100),
                '영어': random.randint(0, 100),
                '수학': random.randint(0, 100)
            }
            student["학생 " + str(stu_num + 1)] = point                         # 학생 key를 생성하고 랜덤 생성된 점수를 value로 저장
        room[str(room_num + 1) + "반"] = student                               # 반 key를 생성하고 학생데이터를  value로 저장
    school[str(grade + 1) + "학년"] = room                                      # 학년 key를 생성하고 반데이터를 value로 저장
                                                                                # 반복에 따라 새로운 데이터를 지속생성하게 해야 중복데이터가 생기지 않음

print(school["1학년"])
print(school["2학년"])
print(school["3학년"])
```

```python
# 학년과 과목으로 해당 학년의 과목 평균값을 구함

ginput = input("학년을 입력하세요:(1~6): ")+"학년"
subinput = input("과목을 입력하세요(국어, 영어, 수학): ")
# 테스트 1
get_grade_average(ginput,subinput,school)                                       # 학년과 과목 값만 주어져 있을 때 함수 실행
# 테스트 2
rinput = input("반을 입력하세요:(1~5): ")+"반"                                  # rinput 값을 추가했을 때 함수 실행
get_grade_average(ginput,subinput,school,rinput)
```

```python
# 과목을 입력받아 모든 학년의 평균값을 구함
# 테스트 3
get_subaverage(school)
```
