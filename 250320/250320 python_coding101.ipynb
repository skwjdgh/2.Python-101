```python
#소수 출력하기

count = 0
prt = [0,0,0,0,0]

while(count < 5):
    val = int(input())
    rcnt = 0
    for a in range(1, val+1):
        if val % a == 0:
            rcnt = rcnt + 1

    if rcnt == 2:
        prt[count]=val
        count = count + 1

print(prt)
```

```python
#교수님 방법
p_list = [None,None,None,None,None]
p_index = 0 #index를 통해 위 리스트의 위치를 가르키기 위한 변수, 0부터 시작해야하기 때문
while p_index < 5: # 반복 횟수를 모르기 때문에 while문 사용, for은 시퀸스가 필요함

    num = int(input("정수를 입력해주세요 :")) #입력값이 전역변수로 사용할 필요가 없음
    num_c = 0 #약수의 개수를 확인하기 위한 변수
    for a in range(1, num+1): #입력 받은 숫자의 약수의 개수를 찾기 --> 나중에 나누었을 때 1과 본인을 제외한 수 중에 나머지가 0인 것이 있는가
        #print(a) #a의 내부값을 확인해보자
        if num % a == 0 :  #입력받은 값과 배열로 생성된 값의 나머지가 0일 경우
            num_c = num_c + 1 #약수의 개수 카운트

    if num_c == 2: #약수의 개수가 2일 경우  #num_c가 왜 여기에 있을까?
        p_list[p_index] = num #p_list의 p_index 값에 맞는 위치에 저장한다
        p_index = p_index + 1 #p_index 값을 증가시킴


print("소수 : ",p_list)
```

```python
def p_check(num): # 디폴트 값을 가지면 무조건 반환
    num_c = 0
    for a in range(1, num+1):
        if num % a == 0 :
            num_c = num_c + 1
    if num_c == 2:
        return True #반환 값 그대로 함수로 적용
    else:
        return False #값을 반환하지 않음

p_list = [None,None,None,None,None]
p_index = 0

while p_index <5 :

    num = int(input("정수를 입력해주세요 :"))
    if p_check(num):
        p_list[p_index] = num #함수에서 반환한 값을 리스트에 입력 -> append 사용가능
        p_index = p_index + 1

print("소수",p_list)
```
