```python
#소수 넣기 코드

print_list = [None,None,None,None,None]
print_index = 0

while True :

    input_int = int(input()) #입력 받기 위한 변수 선언
    num_count = 0 # 연산 후 몫의 개수를 확인하기 위한 변수 선언
    for temp_val in range(1,input_int+1): # temp_val은 리스트 형으로 나타내어 지고 1부터 input_int+1까지의 수를 아이템으로 가지게 됨
        if input_int % temp_val == 0: # 만약에 입력3 값과 임시 저장 값의 나누어진 나머지가 0이라면
            num_count = num_count + 1 #num_count는 num_count + 1이다

    if num_count == 2 : # 위 코드 블럭에서의 num_count 값이 2라면 입력값은 소수라고 할 수 있다.
        print_list[print_index] = input_int
        print_index = print_index + 1

    if print_index == 5:
        break

print("입력 받은 숫자 중 소수에 해당하는 값은: ", print_list, "이다")
```

```python
n = 100
primes = [0] * n
count = 0

for i in range(2, n + 1): #2부터 100까지의 범위
    for j in range(2, i + 1): #2부터 100까지의 범위
        if i % j == 0: #i와 j를 나누었을 때 몫이 0인가
            if j == i: #i와 j가 같을 때
                primes[count] = i
                count = count+1
            break

print(primes[:count]) #슬라이
```
