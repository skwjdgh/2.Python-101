```python
import random
import os

# 투구값을 구하는 함수
def rand(ran):
    rrange = random.randint(0,ran)
    return rrange

    # 함수1 각 스코어에 맞는 점수값을 자료형에 입력
def point(fr,result, tr1, tr2, tr3, st): # 프레임, 결과, 1투구,2투구, 3투구, 상태
    currentframe = str(fr+1)+"프레임"
    result[currentframe][0] = tr1
    result[currentframe][1] = tr2
    result[currentframe][2] = tr3
    result[currentframe][3] = result[currentframe][0] + result[currentframe][1] + result[currentframe][2]

    if fr > 0:
        prevframe = str(fr) + "프레임"
        result[currentframe][4] = result[prevframe][4] + result[currentframe][3]
    else:
        result[currentframe][4] = result[currentframe][3]

    result[currentframe][5] = st
    return result

    # 함수2 이전 프레임의 상태값을 확인해서 투구값을 과거 프레임 결과값에 추가하고 현재 누적값에 추가하는 함수
def pastpoint(fr, result, tr): # tr 현재 투구값
    for pastframes in range(fr-2, fr):
        if pastframes < 0:
            continue
        else:
            pastframe = str(pastframes+1)+"프레임"
            if result[pastframe][5] > 0:
                result[pastframe][3] = result[pastframe][3] + tr
                result[pastframe][5] = result[pastframe][5] - 1
                if pastframes > 0:
                    prev = str(pastframes) + "프레임"  #이전 프레임의 누적값에  이번 프레임의 합계를 더람
                    result[pastframe][4] = result[prev][4] + result[pastframe][3]
                else:
                    result[pastframe][4] = result[pastframe][3]


                    # 볼링게임
def bowlinggame():
    defaultframe = 10
    defaultpin = 10
    normal = 0
    spare = 1
    strike = 2
    throw1 = 0
    throw2 = 0
    throw3 = 0
    sum = 0
    total = 0
    result = {}

    for frames1 in range(defaultframe):
        currentframe = str(frames1+1)+"프레임"
        result[str(frames1+1)+"프레임"] = [0,0,0,0,0,0]


    for frames in range(defaultframe):
        print("{} 프레임".format(frames))
        throw1 = rand(defaultpin)
        print("1 투구는 {}점입니다.".format(throw1))

        if throw1 == defaultpin:
            print("스트라이크")
            point(frames,result,throw1,0,0,strike)
            pastpoint(frames, result, throw1)

        else:
            temppin = defaultpin - throw1
            throw2 = rand(temppin)
            pastpoint(frames, result, throw1)

            if throw2 == temppin:
                print("스페어")
                point(frames,result,throw1,throw2,0,spare)
                pastpoint(frames, result, throw2)

            else:
                print("스페어 실패")
                point(frames,result,throw1,throw2,0,normal)
                pastpoint(frames, result, throw2)

        if frames == 9:
            throw1 = rand(defaultpin)
            pastpoint(frames, result, throw1)
            currentframe = str(10) + "프레임"
            prevframe = str(9) + "프레임"

            if throw1 == defaultpin:

                throw2 = rand(defaultpin)
                pastpoint(frames, result, throw2)

                if throw2 == defaultpin:
                    throw3 = rand(defaultpin)
                    point(frames,result,throw1,throw2,throw3,0)

                else:
                    temppin1 = defaultpin - throw2
                    throw3 = rand(temppin1)
                    point(frames,result,throw1,throw2,throw3,0)

            else:
                temppin = defaultpin - throw1
                throw2 = rand(temppin)
                pastpoint(frames, result, throw2)

                if throw2 == temppin:
                    throw3 = rand(defaultpin)
                    point(frames,result,throw1,throw2,throw3,0)

                else:
                    point(frames,result,throw1,throw2,0,0)

    return result


# 게임 수 만큼 볼링게임을 돌려 위 딕셔너리 자료형을 채움
maxgames = 10
game = random.randint(0,maxgames)
p_list = []
p_list.append(bowlinggame for bowlinggame in range(game))

#만들 파일 개수
pop = ["태조","정조","태종","세종","문종","단종","세조","예종","성종","연산군","중종","인종","명종","선조"]
makefile = random.randint(0,len(pop))  # 인덱스 개수 생성

makefiles = {}
for files in range(makefile):
    makefiles[pop[files]] = p_list

print(makefiles)

```
