```python
# 랜덤 모듈 소환
import random

# 숫자의 크기를 비교하는 함수
def compare_num(start, end, limit, temp1):                  #(시작, 끝, 횟수, 찾는 값)

    count = 0                                               # 시도 횟수 카운터
    result = None                                           # 찾는 값이 나왔을 경우, 결과값 출력

    while(count < limit):                                   # 시도 횟수가 주어진 횟수보다 적을 때 동작

        #temp2 = random.randrange(start,end)                 # 범위 내 만들어진 비교값,시작값과 끝값이 변하며 랜덤 수를 만들어내어 찾는 값temp1과 비교
        temp2 = int(input("숫자를 입력하세요."))
        if temp1 > temp2:                                   # 찾는 값을 랜덤 값과 비교
            print("찾는 값은", temp2,"보다 큽니다.")
            start = temp2                                   # 시작 값을 temp2로 변경
            count = count+1                                 # 시도 횟수 차감

        elif temp1 < temp2:
            print("찾는 값은", temp2,"보다 작습니다.")
            end = temp2                                     # 끝 값을 temp2로 변경
            count = count + 1                               # 시도 횟수 차감

        else:
            result = temp2                                  #result에 temp2 값 초기화
            break                                           #값을 찾았으니 과정 종료

    return count,result                                     # 함수를 통해 구한 결과값과 시도 횟수를 리턴

# 기본적으로 필요한 입력값

start = int(input("시작 값을 입력하세요."))
end = int(input("끝 값을 입력하세요."))
limit = int(input("몇 번까지 시도할지 입력하세요"))

# 해당 범위의 랜덤 값을 select 에 저장하고 함수에 값을 전달해 결과를 출력하는 코드 블록
select = random.randrange(start,end)                        # 입력된 범위 내에서 랜덤 값을 다른 변수에 저장

count, result = compare_num(start,end,limit,select)         # 함수를 통해 구한 결과값과 시도 횟수를 초기화

if result is not None:                                      # 제한 횟수 내에 결과값을 구했을 때
    print("당신은",count ,"번 만에 정답 값을 구했습니다.")
else:                                                       # 제한 횟수 내에 결과값을 못 구했을 때, count출력을 통해 정상 동작 확인
    print("당신은",count ,"의 시도 중 값을 찾지 못했습니다.")

print("찾는 값은", select,"입니다.")
```
