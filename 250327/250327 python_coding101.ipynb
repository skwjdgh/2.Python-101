```python
# 다섯 자리의 숫자를 입력 받고 양수 음수를 판단한 다음, 10개의 양수가 모이면 출력

int_input = []                                                      # 입력 값이 모일 리스트
temp_list =[]                                                       # 출력 값이 모일 리스트

while len(temp_list) < 10:                                          # 양의 정수가 10개 모일 때까지 동작
                                                                    # -> 입력 값을 받기 때문에 출력값이 10자리 이상이 됨
    for temp in range(5):                                           # 반복 문을 통해 입력된 수를 제한

        value = input("정수를 입력하세요: ")                        # 문자 입력
        a = int(value)                                              # 문자를 정수형으로 형변환
        int_input.append(a)                                         # 입력 값을 int_input이라는 리스트에 초기화

        if a >= 0 :                                                 # a값의 부호 판단
            temp_list.append(a)                                     # a값이 0 혹은 양수라면 temp_list라는 리스트에 초기화
        else :
            continue                                                # 다른 값을 가질 경우, 반복문 계속

        if len(temp_list) >= 10:                                    # temp_list의 길이가 10보다 크거나 같으면 루프 "즉시" 종료
            break                                                   # => 10개만 출력하도록 함


    print("지금까지 입력받은 수는", int_input,"입니다.")            # 지금까지 입력값 리스트

print("입력 받은 수 중 10개의 양의 정수는 ",temp_list,"입니다.")    # 10개의 양의 정수 출력
```

```python
# 다섯 자리의 숫자를 입력 받고 양수 음수를 판단한 다음, 10개의 양수가 모이면 출력

int_input = []                                                      # 입력 값이 모일 리스트
temp_list =[]                                                       # 출력 값이 모일 리스트

while len(temp_list) < 10:

    value = list(map(int, input("5개의 정수를 입력하세요.").split()))
    if value != 5:
        continue

    for num in value:
        int_input.append(num)

        if num >= 0 :
            temp_list.append(num)

        if len(temp_list) >= 10:
            break

    print("지금까지 입력받은 수는", int_input,"입니다.")

print("입력 받은 수 중 10개의 양의 정수는 ",temp_list,"입니다.")
```
