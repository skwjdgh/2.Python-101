```python
# 간단한 함수들
# 1. 범위를 구하는 함수

def is_range(frames):
    return range(0,frames)

def is_throw(pins):
    return random.randint(0,pins)
```

```python
# 결과를 모을 자료 생성
result = {f"{frames+1}프레임" : [0,0,0,0,0,0] for frames in is_range(defaultframe)}
```

```python
# 볼링 게임

for frames in is_range(defaultframe):
    currentframe = str(frames+1) + "프레임"
    throw1 = is_throw(defaultpin)


    if throw1 == defaultpin:

        is_pastframe(frames,throw1, result)
        is_total(currentframe, throw1, 0, 0 ,result, strike) #이전값 연산이 먼저 연산되도록?

    else :
        temppin = defaultpin - throw1
        throw2 = is_throw(temppin)

        if temppin+throw2 == defaultpin:

            is_pastframe(frames,throw1, result)
            is_pastframe(frames,throw2,result)
            is_total(currentframe, throw1, 0, 0 ,result, spare)

        else:
            is_pastframe(frames,throw1, result)
            is_pastframe(frames,throw2,result)
            is_total(currentframe, throw1,throw2,0,result,normal)

    if frames == 9:
        throw1 = is_throw(defaultpin)
        is_pastframe(frames,throw1, result)

        if throw1 == defaultpin:
            throw2 = is_throw(defaultpin)
            is_pastframe(frames,throw2,result)

            if throw2 == defaultpin:
                throw3 = is_throw(defaultpin)

            else:
                temppin2 = defaultpin- throw2
                throw3 = is_throw(temppin2)

            is_total(currentframe, throw1, throw2, throw3 ,result, normal)

        else:
            temppin1 = defaultpin - throw1
            throw2 = is_throw(temppin1)
            is_pastframe(frames,throw2,result)

            if temppin1 + throw2 == defaultpin:
                throw3 = is_throw(defaultpin)
                is_total(currentframe, throw1, throw2, throw3 ,result, normal)

            else :
                is_total(currentframe, throw1, throw2, 0 ,result, normal)

print(result)
```
